Escrever um programa que recebe a temperatura em graus celcius e converte para fahrenheit

#Entrada: graus celsius


#Processamento

F = C x (9/5) + 32

#Saída

Mostrar graus em farenheit

```python
#Entrada


celsius = float(input("     Digite a Temperatura em Celsius:   "))


#Processamento


farenheit = 1.8 * celsius + 32


#Saída


print ("|             MAQUINA DE CONVERSÃO                  |")
print ("|                                                   |")
print ("|                                                   |")
print ("|    Conversão De Graus Celsius Para Farenheit      |")
print ("|                                                   |")
print ("|                                                   |")
print(f"|  {celsius} C° convertendo em Farenheit é {farenheit}:          |")

```
